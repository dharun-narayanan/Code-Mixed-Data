<h2>Preprocessing</h2>

```python
corp = f.split("\n")
corp
```

```python
def text_lowercase(text):
    return text.lower()

def remove_numbers(text):
    result = re.sub(r'\d+','',text) #regular expression
    return result

def remove_whitespace(text):
    return " ".join(text.split()) #text.strip()

def remove_punctuation(text):
    translator = str.maketrans('','',string.punctuation)
    return text.translate(translator)
```

```python
corpora = [""]*len(corp)
for i in range(len(corp)):
    raw = text_lowercase(corp[i])
    processed = remove_whitespace(raw)
    processed = remove_numbers(processed)
    processed = text_lowercase(processed)
    processed = remove_punctuation(processed)
    processed = processed.strip()
    processed = "<s> "+processed+" </s>"
    corp[i] = processed
    corpora[i] = corp[i][4:-5]   
```

```python
corp
```

```python
tokens = []
for i in range(len(corpora)):
    word = word_tokenize(corpora[i])
    tokens = tokens+word
tokens
```

<b>Separating English and Tamil Words</b>

```python
import nltk
words = set(nltk.corpus.words.words())

eng = " ".join(w for w in tokens \
         if w.lower() in words or not w.isalpha())
eng
```

```python
eng = word_tokenize(eng)
eng
```

```python
for i in eng:
    if len(i)<=2:
        eng.remove(i)
eng
```

```python
word_tokens = eng
tagged = nltk.pos_tag(word_tokens)
print("POS Tagging: \n")
print(tagged)
```

```python
tam = [x for x in tokens if x not in eng]
tam
```

```python
tokenized_text  = []
for i in range(len(corpora)):
    word = word_tokenize(corpora[i])
    tokenized_text.append(word)
tokenized_text 
```

<h3>Translation</h3>

```python
translated_corp = ""
for sent in tokenized_text:
    for word in sent:
        if word in tam:
            suggestions = transliterate_word(word, lang_code='ta')
            tran = translator.translate(str(suggestions[0])).text
            translated_corp = translated_corp+" "+tran
        else:
            translated_corp = translated_corp+" "+str(word)
    translated_corp = translated_corp+" | "
translated_corp 
```
